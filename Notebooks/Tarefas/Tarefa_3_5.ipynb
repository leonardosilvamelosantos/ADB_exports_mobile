#### c-) Mostrar os histogramas dos atributos: *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
import pandas as pd
import matplotlib.pyplot as plt


HW = pd.read_csv('DSBD/Datasets/Hourly_wages/hourly_wages.csv', na_values=['?'])


fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].hist(HW['education_yrs'], bins=10, color='blue', alpha=0.7)
axs[0].set_title('Education Years')
axs[0].set_xlabel('Years')
axs[0].set_ylabel('Frequency')

axs[1].hist(HW['experience_yrs'], bins=10, color='green', alpha=0.7)
axs[1].set_title('Experience Years')
axs[1].set_xlabel('Years')
axs[1].set_ylabel('Frequency')

axs[2].hist(HW['age'], bins=10, color='red', alpha=0.7)
axs[2].set_title('Age')
axs[2].set_xlabel('Age')
axs[2].set_ylabel('Frequency')

# Ajuste o espaçamento entre os subplots
plt.tight_layout()

# Exiba o gráfico
plt.show()
```

#### d-) Mostrar os gráficos de setores (*pie chart*) de cada um dos atributos: *union*, *female*, *marr*, *south*, *manufacturing* e *construction* num mesmo quadro de plotagem (usando subplots).

```

atributos = ['union', 'female', 'marr', 'south', 'manufacturing', 'construction']

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, atributo in enumerate(atributos):
    row = i // 3
    col = i % 3
    labels = HW[atributo].value_counts().index
    sizes = HW[atributo].value_counts().values
    axs[row, col].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[row, col].set_title(atributo)

plt.tight_layout()

plt.show()
```

#### e-) Mostrar os boxplots das variáveis *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
variaveis = ['education_yrs', 'experience_yrs', 'age']

# Crie uma figura com subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Itere sobre as variáveis e crie boxplots para cada uma
for i, variavel in enumerate(variaveis):
    axs[i].boxplot(HW[variavel], vert=False)
    axs[i].set_title(variavel)
    axs[i].set_xlabel('Valores')

# Ajuste o layout dos subplots
plt.tight_layout()

# Exiba o gráfico
plt.show()
```

#### f-) Mostre lado a lado os boxplots da variável *age* discriminados por sexo (female).

```
feminino = HW[HW['female'] == 1]
masculino = HW[HW['female'] == 0]

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].boxplot(feminino['age'], vert=False)
axs[0].set_title('Boxplot da Idade - Feminino')
axs[0].set_xlabel('Idade')
axs[0].set_yticklabels(['Feminino'])

axs[1].boxplot(masculino['age'], vert=False)
axs[1].set_title('Boxplot da Idade - Masculino')
axs[1].set_xlabel('Idade')
axs[1].set_yticklabels(['Masculino'])

plt.tight_layout()

plt.show()
```

#### j-) Mostre lado a lado os boxplots da variável wage_per_hour discriminados por estado civil (marr).
